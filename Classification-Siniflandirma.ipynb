## Modelling

```python
#Attrition yani isten cikma kolonumuzda 265 tane bos veri var, bunu doldurmak istiyoruz.
```

```python
df['Attrition']=df['Attrition'].replace(['Yes', 'No'],[1,0])

#dictionary=d=sozluk
#d={"Yes":1,"No":0}
#df["Attrition"]=df["Attrition"].map(d)   #yukaridaki kodla ayni islevi goruyor
#dfnin yeni degeri = dfnin eski degeri [yes ve no yu bulup 1 ve 0 la degistirecek]
```

```python
#Attrition kolonundaki Yesleri 1 Nolari 0 a atadik. Artik tahminlerimizi sayisala dokup 1 ve 0 alicaz
```

```python
bos=df[df["Attrition"].isnull()]   #bos verileri bos diye bir degiskene atadik
```

```python
bos.shape   #265 tane bos veri var
```

```python
bos
```

```python
dolu=df[df["Attrition"].notnull()]   #bos olmayan dolu olan verileri de ayiriyoruz, egitim icin
```

```python
dolu.shape   #1205 dolu veri var
```

```python
dolu
```

```python
#1205 tane veriyi modele egitecegiz, daha sonra ogrenecek makine
#ogrenecegi bilgilerle 265 kisinin isten ayrilip ayrilmayacagini tahmin etirecegiz.
```

```python
x=dolu.drop("Attrition", axis=1)   #x verisinden yani egitecegimiz veriden Cevaplarin oldugu kolonu sildik ve x e atadik
y=dolu["Attrition"]   #gercek cevaplarin oldugu kolonu ise y diye bir degiskene atadik. bununla karsilastirma yapacak
```

```python
x=pd.get_dummies(x,drop_first=True)  
#data turu object olan/yazi olan satirlari sutune donusturup numerik deger verdik
#object olan butun kolonladaki degiskenleri sutun yapti ve uint8 data turune cevirdi
#bruada 3 adet yazi sutunumuz var (BusinessTravel,Department,EducationFIeld)
#bunlari rakama ceviriyoruz. one-hot encoding ile yapiyoruz
#artik verimiz modele hazir, GaussinNB,BerrnoulliNB hepsini cagirabiliriz
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
g=GaussianNB()
b=BernoulliNB()

#sklearn kutuphanesinden cesitli clasification islemlerini yapabilmek icin
#GaussianNB ve BernoulliNB modellerini import ettik.
#bunlar bizim verilerimizi siniflandirmamiza yariyor.
```

```python
g.fit(x,y) #2 modelde egitildi fit komutuyla
```

```python
b.fit(x,y)   #2 modelde egitildi fit komutuyla
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
#basari oranlarimiza bakmak icin gerekli metrikleri import ediyoruz
```

```python
tahmin1=g.predict(x)
tahmin2=b.predict(x)
#2 modelde predict komutu ile tahmin edildi
#gaussianNB modeli ile tahmin ettiklerimiz tahmin 1 e, BernoulliNB ile tahmin edilenler tahmin 2 ye
```

```python
tahmin1
```
